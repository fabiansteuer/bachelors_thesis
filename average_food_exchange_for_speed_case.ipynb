```python
%pylab inline --no-import-all

import os
import time
from GraphSeries import GraphSeries
import matplotlib.pyplot as plt
import numpy as np
#import networkx as nx

from matplotlib import rc
rc('font',
   **{#'family':'sans-serif',
      #'sans-serif':['sans-serif'],
      'family':'serif',
      'serif':['Palatino'],
      'size':22
      })
rc('text',
   usetex=True)
rc('figure',
   **{'figsize':(16,9),
      'dpi':300}
   )

path = '../data/speed/'

folders = sorted([folder
                  for folder in os.listdir(path)
                  if folder.isdigit()
                  ],
                  key=lambda x: int(x)
                  )

factors = []
values = []

for folder in folders:
    #if int(folder) in [20, 30, 40, 50, 80, 100]: continue
    
    with open(path + folder + '/amounts_exchanged.txt') as file:
        data = file.readline().split()
        mean = float(data[0])
    with open(path + folder + '/emptytime.txt') as file:
        data = file.readline().split()
        time = float(data[0])
        
    factors.append(int(folder))
    values.append(mean/time)
             

for n in range(len(factors)):      
    plt.plot(factors[n], values[n], 'o', label=factors[n])
        
# Adjust axes limits
xlims = list(plt.gca().get_xlim())
xlims[0] = xlims[0] - (xlims[1]-xlims[0])*0.01*9/16
xlims[1] = xlims[1] + (xlims[1]-xlims[0])*0.01*9/16
plt.xlim(xlims)
ylims = list(plt.gca().get_ylim())
ylims[0] = ylims[0] - (ylims[1]-ylims[0])*0.01
ylims[1] = ylims[1] + (ylims[1]-ylims[0])*0.01
plt.ylim(ylims)

plt.xlabel(r'Speed factor')
plt.ylabel(r'$\frac{\textrm{Total amount of food exchanged}}{\textrm{Runtime of the simulation}}$')
#plt.legend(title='Speed factor')
plt.tight_layout()
plt.savefig('../data/speed/amount-runtime-ratio.pdf')
```
