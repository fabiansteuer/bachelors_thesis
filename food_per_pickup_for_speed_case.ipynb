```python
%pylab inline --no-import-all

import os
import time
import numpy as np
from GraphSeries import GraphSeries
import matplotlib.pyplot as plt
#import networkx as nx

from matplotlib import rc
rc('font',
   **{#'family':'sans-serif',
      #'sans-serif':['sans-serif'],
      'family':'serif',
      'serif':['Palatino'],
      'size':22
      })
rc('text',
   usetex=True)
rc('figure',
   **{'figsize':(16,9),
      'dpi':300}
   )

path = '../data/speed/'

folders = sorted([folder
                  for folder in os.listdir(path)
                  if folder.isdigit()
                  ],
                  key=lambda x: int(x)
                  )
for folder in folders:
    with open(path + folder + '/pickups_exact.txt') as file:
        data = file.readline().split()
        print(data)
        if int(folder) in [80, 100]: continue
        #plt.gca().set_yscale('log')
        #plt.errorbar(int(folder), 112/float(data[0]), yerr=112*float(data[1])/float(data[0])**2, fmt='o', linewidth=3, label=str(folder))
        plt.plot(int(folder), float(data[0]), 'o', markersize=10)
        plt.xlabel(r'Speed factor')
        plt.ylabel(r'$\frac{\textrm{Amount of food picked up per pickup}}{\textrm{Food capacity of one ant}}$')
        #plt.legend(title='Step factor', numpoints=1)
        
# Adjust axes limits
xlims = list(plt.gca().get_xlim())
xlims[0] = xlims[0] - (xlims[1]-xlims[0])*0.01*9/16
xlims[1] = xlims[1] + (xlims[1]-xlims[0])*0.01*9/16
plt.xlim(xlims)
ylims = list(plt.gca().get_ylim())
ylims[0] = ylims[0] - (ylims[1]-ylims[0])*0.01
ylims[1] = ylims[1] + (ylims[1]-ylims[0])*0.01
plt.ylim(ylims)

plt.tight_layout()
plt.savefig(path + 'amount_per_pickup.pdf') 
```
