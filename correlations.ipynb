```python
%pylab inline --no-import-all

import os
import time
import numpy as np
from GraphSeries import GraphSeries
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
#import networkx as nx

from matplotlib import rc
rc('font',
   **{#'family':'sans-serif',
      #'sans-serif':['sans-serif'],
      'family':'serif',
      'serif':['Palatino'],
      'size':22
      })
rc('text',
   usetex=True)
rc('figure',
   **{'figsize':(16,9),
      'dpi':300}
   )

path = '../data/speed/'

folders = sorted([folder
                  for folder in os.listdir(path)
                  if folder.isdigit()
                  ],
                  key=lambda x: int(x)
                  )

emptytimes = []
pickups = []
interactions = []
amountsex = []
for folder in folders:
    if int(folder in [80, 100]): continue
    with open(path + folder + '/emptytime.txt') as efile:
        with open(path + folder + '/pickups.txt') as pfile:
            with open(path + folder + '/interactions.txt') as afile:
                with open(path + folder + '/amounts_exchanged.txt') as aefile:
                    edata = efile.readline().split()
                    pdata = pfile.readline().split()
                    adata = afile.readline().split()
                    aedata = aefile.readline().split()
                    #print(data)
                    #if int(folder) in [80, 100]: continue
                    #plt.gca().set_yscale('log')
                    #plt.errorbar(int(folder), (float(edata[0])/float(pdata[0])), yerr=float(edata[1])/1000, fmt='o', linewidth=3, label=str(folder))
                    #plt.xlabel(r'Speed factor')
                    #plt.ylabel(r'Ratio')
                    #plt.legend(title='Step factor', numpoints=1)
                    emptytimes.append(float(edata[0]))
                    pickups.append(float(pdata[0]))
                    interactions.append(float(adata[0]))
                    amountsex.append(float(aedata[0]))
print('Runtime-Number of Pickups:')            
print(pearsonr(emptytimes, pickups))
print('Runtime-Amount-per-pickup:')
print(pearsonr(emptytimes, 112/np.array(pickups)))
print('Runtime-Amount-per-interaction:')
print(pearsonr(emptytimes, np.array(amountsex)/np.array(interactions)))
            
# Adjust axes limits
#xlims = list(plt.gca().get_xlim())
#xlims[0] = xlims[0] - (xlims[1]-xlims[0])*0.01*9/16
#xlims[1] = xlims[1] + (xlims[1]-xlims[0])*0.01*9/16
#plt.xlim(xlims)
#ylims = list(plt.gca().get_ylim())
#ylims[0] = ylims[0] - (ylims[1]-ylims[0])*0.01
#ylims[1] = ylims[1] + (ylims[1]-ylims[0])*0.01
#plt.ylim(ylims)
```
