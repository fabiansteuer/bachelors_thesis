```python
%pylab inline --no-import-all

import os
import time
from GraphSeries import GraphSeries
import matplotlib.pyplot as plt
import numpy as np
#import networkx as nx

from matplotlib import rc
rc('font',
   **{#'family':'sans-serif',
      #'sans-serif':['sans-serif'],
      'family':'serif',
      'serif':['Palatino'],
      'size':22
      })
rc('text',
   usetex=True)
rc('figure',
   **{'figsize':(16,9),
      'dpi':300}
   )

path = '../data/speed/'

folders = sorted([folder
                  for folder in os.listdir(path)
                  if folder.isdigit()
                  ],
                  key=lambda x: int(x)
                  )
def gaus(x, a, x0, sigma):
            return a*np.exp(-(x-x0)**2/(2*sigma**2))

for folder in folders:
    if int(folder) in [20, 30, 40, 50, 80, 100]: continue
    with open(path + folder + '/gaus.txt') as file:
        data = file.readline().split()
        a = float(data[0])
        mean = float(data[1])
        std = float(data[2])
        print(data) 
        
        x = np.linspace(mean-3*std, mean+3*std, 100)
        plt.plot(x, gaus(x, a, mean, std), linewidth=2, label=folder)
        
# Adjust axes limits
xlims = list(plt.gca().get_xlim())
xlims[0] = xlims[0] - (xlims[1]-xlims[0])*0.01*9/16
xlims[1] = xlims[1] + (xlims[1]-xlims[0])*0.01*9/16
plt.xlim(xlims)
ylims = list(plt.gca().get_ylim())
ylims[0] = ylims[0] - (ylims[1]-ylims[0])*0.01
ylims[1] = ylims[1] + (ylims[1]-ylims[0])*0.01
plt.ylim(ylims)

plt.xlabel(r'Number of interactions')
plt.ylabel(r'Average number of ants per run')
plt.legend(title='Speed factor')
plt.tight_layout()
plt.savefig('../data/speed/gaussians.pdf')
```
