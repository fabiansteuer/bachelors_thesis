```python
%pylab inline --no-import-all

import os
import time
from GraphSeries import GraphSeries
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt

#import networkx as nx

from matplotlib import rc
rc('font',
   **{#'family':'sans-serif',
      #'sans-serif':['sans-serif'],
      'family':'serif',
      'serif':['Palatino'],
      'size':22
      })
rc('text',
   usetex=True)
rc('figure',
   **{'figsize':(16,9),
      'dpi':300}
   )

path = '../data/speed/'

folders = sorted([folder
                  for folder in os.listdir(path)
                  if folder.isdigit()
                  ],
                  key=lambda x: int(x)
                  )

factors = []
values = []
valueerrors = []
twos = []
twoerrors = []

for folder in folders:
    if int(folder) in [80, 100]: continue
    
    with open(path + folder + '/amounts_exchanged.txt') as file:
        data = file.readline().split()
        mean = float(data[0])
        meanerror = float(data[1])
    with open(path + folder + '/interactions.txt') as file:
        data = file.readline().split()
        time = float(data[0])
        timeerror = float(data[1])
    
        
    factors.append(int(folder))
    values.append(mean/time)
    valueerrors.append(sqrt(meanerror**2/time**2 + timeerror**2*mean**2/time**4))
    
    with open(path + folder + '/amount_per_interaction.txt') as file:
        data = file.readline().split()
        two = float(data[0])
        twoerror = float(data[1])
    twos.append(two)
    twoerrors.append(twoerror)

for n in range(len(factors)):      
    #plt.errorbar(factors[n], values[n], yerr=valueerrors[n]) #label=factors[n])
    plt.plot(factors[n], values[n], 'o', markersize=10) #label=factors[n])
    
#plt.figure()
#for n in range(len(factors)):      
#    plt.errorbar(factors[n], twos[n], yerr=twoerrors[n]) #label=factors[n])
#    plt.plot(factors[n], twos[n], 'o') #label=factors[n])
        
# Adjust axes limits
xlims = list(plt.gca().get_xlim())
xlims[0] = xlims[0] - (xlims[1]-xlims[0])*0.01*9/16
xlims[1] = xlims[1] + (xlims[1]-xlims[0])*0.01*9/16
plt.xlim(xlims)
ylims = list(plt.gca().get_ylim())
ylims[0] = ylims[0] - (ylims[1]-ylims[0])*0.01
ylims[1] = ylims[1] + (ylims[1]-ylims[0])*0.01
plt.ylim(ylims)

plt.xlabel(r'Speed factor')
plt.ylabel(r'$\frac{\textrm{Average food exchange per interaction}}{\textrm{Food capacity of one ant}}$')
#plt.legend(title='Speed factor')
plt.tight_layout()
plt.savefig('../data/speed/amount_per_interaction.pdf')
```
